```python
"""
Sean Lubner

LabVIEW 3-Omega Data File Reader 
"""

# Future Work:
# TODO: lock x-axes of subplots together
# TODO: subtract out individual noisy data points from plot fitting
# TODO: mouse-over data points to show time-series, read in full data file (so points are 
#        selectable), maybe "simple vs non-simple mode"
# TODO: auto-detect if output files already exists, and if so, increment name of new 
#        one by 1 (or have option to over-write pre-existing)
# TODO: error analysis

# Web-App or GUI interface:
# TODO: instead of auto-selecting sensors_data file and sensor, and data file directories, all
#       click selected
# TODO: When user is done using server, delete all objects to clear up memory

#==============================================================================
# Parameter that this function takes in
sensors_file = "SensorsData3w" # file storing calibration files for all active sensors / samples
sensor_name = "Sensor 6 B" #"SiGe 1" #"Liver Sensor 3" # Which sensor to be used for this fit
current_correction = 0.936 # correction factor for the 3-omega voltage-current transducer circuit
#==============================================================================

#==============================================================================
# Import Packages:
import numpy as np
import string
import os
import matplotlib as mpl
import matplotlib.pyplot as plt

from Tkinter import Tk # for file browsing
from tkFileDialog import askopenfilename # for file browsing
    
%matplotlib

# Also used (immportd within functins): os, pickle

#==============================================================================

#==============================================================================
# Classes
#==============================================================================
class data_file(object):
    """
    Data file object, has the following attributes:
    fileName; sensor; figure; comment; length; current; resistance; freq; v1wx; v1wx_std; v1wy; v1wy_std; v3wx; 
    v3wx_std; v3wy; v3wy_std; v1wr; v1wr_std; v3wr; v3wr_std
    """
    def __init__(self, file_name):
        """ Parse the data file and initialize a data file object with the data """
        print 'Select the data file to read-in and plot (see file browser).'
        #browser = Tk()
        #browser.withdraw() # we don't want a full GUI, so keep the root window from appearing
        #self.fileName = askopenfilename() # show an "Open" dialog box and return the path to the selected file
        #browser.destroy() # get rid of the widget window
        self.fileName = file_name # full path for data file
        self.sensors = get_sensors_list(sensors_file)
        self.sensor = self.sensors[sensor_name]
        main_file = self.fileName.split('/')[-1]
        print 'You selected file "' + main_file + '" to plot. \n'
        data_array = np.loadtxt(self.fileName, skiprows=4) # skip date/time, comment, current, headers
        if data_array[0,1] > data_array[-1,1]: # first freq > last freq
            data_array = data_array[::-1] # flip up-down to start at low frequency
        # Organize data    
        self.length, self.width = np.shape(data_array)
        self.freq = data_array[:,1]
        self.v1wx = data_array[:,3]
        self.v1wx_std = data_array[:,4]
        self.v1wy = data_array[:,5]
        self.v1wy_std = data_array[:,6]
        self.v3wx = data_array[:,7]
        self.v3wx_std = data_array[:,8]
        self.v3wy = data_array[:,9]
        self.v3wy_std = data_array[:,10]
        
        text = open(self.fileName).readline().split("\r")
        self.comment = text[1]
        self.current = float(text[-4].split()[-1])*current_correction/1000 # current_correction for imperfect V-I transduction
        self.resistance = self.get_v1wr().mean()/self.current # Averaged over all data
        
        self.fits = [] # initialize an empty list of fits
    def add_fit(self):
        """ Add (and implicitly plot) a fit for this data """
        tempFit = fit(self)
        self.fits.append(tempFit)
    def plot_data(self, selectFit=True):
        """
        Plot V1w and V3w, and have user select frequency interval from which
        to calculate k, if selectFit = True
        trial is a data file formated by data_file()
        """
        print "See plotted figure."
        print ""
        fileTitle = " ".join(self.get_fileName().split("_")[2:])
        self.fig = plt.figure('V1w and V3w: '+fileTitle, facecolor='w', edgecolor='k', figsize=(16,7))
        self.fig.suptitle("Trial: " + self.get_fileName())
        self.fig.subplots_adjust(wspace=0.5, bottom=0.2, top=0.8) # buffer space below x-label and botom of figue window
        
        # Plot 1-omega data
        self.ax1 = self.fig.add_subplot(121)
        self.ax1.errorbar(self.get_freq(),self.get_v1wx(), yerr=self.get_v1wx_std(), capsize=2, 
                     ecolor='k', fmt='bo-', label="In-Phase", linewidth=1) # v1wx
        self.ax1.errorbar(self.get_freq(),self.get_v1wy(), yerr=self.get_v1wy_std(), capsize=2, 
                     ecolor='k', fmt='ro-', label="Out-of-Phase", linewidth=1) # v1wy
        self.ax1.autoscale(axis='both', tight=True) # shrink axes tight around data, then...
        ylimsTemp = plt.ylim() # store limits of y-axis
        pad = abs(ylimsTemp[1]-ylimsTemp[0])*0.1
        plt.ylim((ylimsTemp[0]-pad, ylimsTemp[1]+pad)) # Pad the y-axis range by 10% top & bottom
        self.ax1.legend(loc="best")
        self.ax1.set_title("V$_{1\omega}$ Data: "+fileTitle, fontsize=18)
        self.ax1.set_xlabel("Frequency, $\omega$ [Hz]", fontsize=16)
        self.ax1.set_xscale('log')
        self.ax1.set_ylabel("First Harmonic Voltage, V$_{1\omega}$ [Volts]", fontsize=16)
        self.ax1.tick_params(axis='both', labelsize=12)
        self.ax1.grid(True)
            
        # Plot 3-omega data
        self.ax2 = self.fig.add_subplot(122)
        self.ax2.errorbar(self.get_freq(),self.get_v3wx(), yerr=self.get_v3wx_std(), capsize=2, 
                     ecolor='k', fmt='co-', label="In-Phase", linewidth=1) # v3wx
        self.ax2.errorbar(self.get_freq(),self.get_v3wy(), yerr=self.get_v3wy_std(), capsize=2, 
                     ecolor='k', fmt='mo-', label="Out-of-Phase", linewidth=1) # v3wy
        self.ax2.autoscale(axis='both', tight=True) # shrink axes to be tight around data, then...
        ylimsTemp = plt.ylim() # store limits of y-axis
        pad = abs(ylimsTemp[1]-ylimsTemp[0])*0.1
        plt.ylim((ylimsTemp[0]-pad, ylimsTemp[1]+pad)) # Pad the y-axis range by 10% top & bottom
        self.ax2.legend(loc="best")
        self.ax2.set_title("V$_{3\omega}$ Data: "+fileTitle, fontsize=18)
        self.ax2.set_xlabel("Frequency, $\omega$ [Hz]", fontsize=16)
        self.ax2.set_xscale('log')
        self.ax2.set_ylabel("Third Harmonic Voltage, V$_{3\omega}$ [Volts]", fontsize=16)
        self.ax2.tick_params(axis='both', labelsize=12)
        self.ax2.grid(True)    
            
        if selectFit: # Perform a linear regression fit to selected region of data and plot / calc props
            self.add_fit()

    # Set properties
    def set_figure(self, figure): # set a reference to the figure being used to plot the data for this file
        self.fig = figure        
        
    # get properties (keep modular and flexible for future modifications)
    def get_fits(self):
        return self.fits
    def get_figure(self): # return reference to the figure object plotting this file's data
        return self.fig
    def get_fileName(self): # return (just the) name of the raw data file for this data file object 
        return self.fileName.split('/')[-1] # remove the path from the file name
    def get_full_fileName(self): # return name of (full path for) the raw data file for this data file object 
        return self.fileName
    def get_sensor(self): # return sensor object used in analyzing fits for this data file
        return self.sensor
    def get_length(self): # return the number of data points
        return self.length
    def get_freq(self): # return the array of frequencies [Hz]
        return self.freq
    def get_v1wx(self): # return the array of in-phase 1-omega voltages [Volts]
        return self.v1wx
    def get_v1wx_std(self): # return the array of in-phase 1-omega voltage standard deviatons [Volts]
        return self.v1wx_std
    def get_v1wy(self): # return the array of out-of-phase 1-omega voltages [Volts]
        return self.v1wy
    def get_v1wy_std(self): # return the array of out-of-phase 1-omega voltage standard deviatons [Volts]
        return self.v1wy_std
    def get_v1wr(self): # return the array of 1-omega voltage magnitudes [Volts]
        return np.sqrt(self.v1wx**2+self.v1wy**2)
    def get_v1wr_std(self): # return the array of 1-omega voltage magnitude standard deviations [Volts]
        return np.sqrt(self.v1wx_std**2+self.v1wy_std**2)
    def get_v3wx(self): # return the array of 3-omega in-phase voltages [Volts]
        return self.v3wx
    def get_v3wx_std(self):  # return the array of 3-omega in-phase voltage standard deviations [Volts]
        return self.v3wx_std
    def get_v3wy(self):  # return the array of 3-omega out-of-phase voltages [Volts]
        return self.v3wy
    def get_v3wy_std(self): # return the array of 3-omega out-of-phase voltage standard deviations [Volts]
        return self.v3wy_std
    def get_v3wr(self): # return the array of 3-omega voltage magnitudes [Volts]
        return np.sqrt(self.v3wx**2+self.v3wy**2)
    def get_v3wr_std(self): # return the array of 3-omega voltage magnitude standard deviations [Volts]
        return np.sqrt(self.v1wx_std**2+self.v1wy_std**2)
    def get_comment(self): # return the comment string saved with the data file
        return self.comment
    def get_current(self): # return the current value used for this trial (float) [Amps]
        return self.current
    def get_resistance(self): # return the calculated electrical resistance [Ohms]
        return self.resistance

class data_selector(object):
    """ Object for interactively selecting and excluding data points for fit """
    def __init__(self, fitter):
        self.fitter = fitter
        self.fig = self.fitter.fig
        self.cid_mouse = self.fig.canvas.mpl_connect('button_press_event', self.onclick)
        self.cid_key_down = self.fig.canvas.mpl_connect('key_press_event', self.onpress)
        self.picks = [] # selected min/max frequencies
        self.unpicks = [] # specific frequencies to exclude
        self.lines_1w = []
        self.lines_3w = []
        self.dots_1w = []
        self.dots_3w = []
        self.picking = True # currently picking freq interval or exclusion points?
    def onclick(self, event): 
        freq_pick = min(self.fitter.base.get_freq(), key=lambda x: abs(x-event.xdata)) # freq point closest to click
        if self.picking:
            if len(self.picks) < 2: # loop around if re-picking interval
                self.picks.append(freq_pick)
                self.picks.sort()
            else:
                self.picks = [freq_pick]
            self.replot_picks()
        else:
            if freq_pick in self.unpicks: 
                self.unpicks.remove(freq_pick) # un-exclude the selected data point
            else:
                self.unpicks.append(freq_pick) # add the selected data point to exclusion list
                self.unpicks.sort()
            self.replot_unpicks()       
    def onpress(self, event):
        if event.key == "t":
            self.picking = not self.picking
        if event.key == "enter":
            self.fig.canvas.mpl_disconnect(self.cid_key_down)
            self.fig.canvas.mpl_disconnect(self.cid_mouse)
    def replot_picks(self):
        plt.figure(self.fig.number) # select the figure in which this data is plotted
        
        for line in self.lines_1w: # get rid of old selection
            line.set_visible(False)
        self.lines_1w=[]
        self.fig.add_subplot(1,2,1) # select the V1w sub-plot
        for pick in self.picks: # plot and store new selection
            temp_line = plt.axvline(x=pick,color='k', linewidth=1) # show fit region
            self.lines_1w.append(temp_line)
        
        for line in self.lines_3w: # get rid of old selection
            line.set_visible(False)
        self.lines_3w=[]
        self.fig.add_subplot(1,2,2) # select the V1w sub-plot
        for pick in self.picks: # plot and store new selection
            temp_line = plt.axvline(x=pick,color='k', linewidth=1) # show fit region
            self.lines_3w.append(temp_line)    
    def replot_unpicks(self):
        plt.figure(self.fig.number) # select the figure in which this data is plotted
        
        for dot in self.dots_1w: # get rid of old selection
            dot.set_visible(False)
        self.dots_1w=[]
        self.fig.add_subplot(1,2,1) # select the V1w sub-plot
        for unpick in self.unpicks: # plot and store new selection
            temp_dot1 = plt.scatter(unpick, self.fitter.base.get_v1wx()[self.fitter.base.get_freq()==unpick], 
                                    s=80, c='r', marker='x', linewidth=2)
            self.dots_1w.append(temp_dot1)
            temp_dot2 = plt.scatter(unpick, self.fitter.base.get_v1wy()[self.fitter.base.get_freq()==unpick], 
                                    s=80, c='r', marker='x', linewidth=2)
            self.dots_1w.append(temp_dot2)
        plt.xlim(self.fitter.xlims_1w) # reset plot axes to original limits
        plt.ylim(self.fitter.ylims_1w)
        
        for dot in self.dots_3w: # get rid of old selection
            dot.set_visible(False)
        self.dots_3w=[]
        self.fig.add_subplot(1,2,2) # select the V1w sub-plot
        for unpick in self.unpicks: # plot and store new selection
            temp_dot1 = plt.scatter(unpick, self.fitter.base.get_v3wx()[self.fitter.base.get_freq()==unpick], 
                                    s=80, c='r', marker='x', linewidth=2)
            self.dots_3w.append(temp_dot1)
            temp_dot2 = plt.scatter(unpick, self.fitter.base.get_v3wy()[self.fitter.base.get_freq()==unpick], 
                                    s=80, c='r', marker='x', linewidth=2)
            self.dots_3w.append(temp_dot2)
        plt.xlim(self.fitter.xlims_3w) # reset plot axes to original limits
        plt.ylim(self.fitter.ylims_3w)
        
    def get_picks(self):
        return self.picks
    def get_unpicks(self):
        return self.unpicks
    
class fit(object):
    """ 
    Fit object, to perform a fit to a data file object "trial"; has the following attributes:
    min_freq_index; max_freq_index; fitted_freq; fitted_v3wx_exp; fitted_v3wy_exp;
    line_fit; fitted_v3wx_calc; fitted_v3wy_calc; R_squared
    """
    def __init__(self, trial, add_fit_to_plot=True, save_figs=True):
        self.base = trial # Cross-link this fit with the base experiment trial data to which it's applied
        self.fig = self.base.get_figure() # store figure association for this plot in temp variable
        self.fig.add_subplot(1,2,1) # select the V1w sub-plot
        self.xlims_1w = plt.xlim() # store limits of x-axis
        self.ylims_1w = plt.ylim() # store limits of y-axis
        self.fig.add_subplot(1,2,2) # select the V3w sub-plot
        self.xlims_3w = plt.xlim() # store limits of x-axis
        self.ylims_3w = plt.ylim() # store limits of y-axis
        self.fit_num = len(self.base.get_fits())+1 # record which number fit it is
        plt.figure(self.base.get_figure().number) # select the current figure
        instructionsText = "Select the frequency interval to which you want to fit by clicking bounding points in one of the plots.\nRight click to undo a point selection.  Press enter when done." # tell the user what to do
        instructions = plt.figtext(0.1, 0.87, instructionsText, color="red", fontsize=16, 
                                   bbox=dict(facecolor="yellow", alpha=0.4, edgecolor="black", pad=10)) 
        freqs = self.base.get_freq() # get vector of frequencies
        fit_range = []
        selector = data_selector(self)
        while len(fit_range) != 2: # make sure get min & max freq 
            fit_range = np.sort(map(lambda x: x[0], plt.ginput(n=0, timeout=0, show_clicks=False)))
            fit_range = selector.get_picks()
            if len(fit_range) < 2:
                instructions.set_text('You must select two points to define a frequency interval. Please try again.')
                plt.draw()
        instructions.set_visible(False) # get rid of instructions
        
        # Calculate indices for points to use for fitting
        self.min_freq_index = list(freqs).index(selector.get_picks()[0])
        self.max_freq_index = list(freqs).index(selector.get_picks()[1])+1
        freqs_picked = freqs[self.min_freq_index:self.max_freq_index] # within the selection window
        self.fit_indices = [i+self.min_freq_index for i, num in enumerate(freqs_picked) if num not in selector.get_unpicks()]
        
        # Calculate fitting quantities
        self.fitted_freq = freqs[self.fit_indices] # fitted frequency interval
        self.fitted_v3wx_exp = self.base.get_v3wx()[self.fit_indices] # fitted experimental v3wx data interval
        self.fitted_v3wy_exp = self.base.get_v3wy()[self.fit_indices] # fitted experimental v3wy data interval
        self.line_fit = np.polyfit(np.log(self.fitted_freq),self.fitted_v3wx_exp,1) # coefficients of best fit line [slope, intercept]
        self.fitted_v3wx_calc = np.polyval(self.line_fit,np.log(self.fitted_freq)) # generate fit points, v3wx [Volts]
        self.fitted_v3wy_calc = np.mean(self.fitted_v3wy_exp) # average out of phase v3wy value [Volts]
        self.fitted_v1wr_exp = self.base.get_v1wr()[self.fit_indices] # fitted experimental v1wr data interval
        self.fitted_v1wr_calc = np.mean(self.fitted_v1wr_exp) # average v1w magnitude value [Volts]
        
        # Calculate R-squared
        SSresiduals = sum((self.fitted_v3wx_exp-self.fitted_v3wx_calc)**2) # sum of squares of residuals
        SStotal = sum((self.fitted_v3wx_exp-np.mean(self.fitted_v3wx_exp))**2) # total sum of squares
        self.R_squared = 1.0 - SSresiduals/SStotal # normalize Rsquared
        
        """ Analyze the 3-omega data file located at /directory/trial.  Parses data file, plots, applies fit, and saves
            a pickle, and both a verbose and compact output string of the results """

        # Calculate / collect properties of interest
        self.resistance = self.fitted_v1wr_calc.mean()/self.base.get_current()
        TCR = self.base.get_sensor().get_RT_fit() # TCR for sensor
        L = self.base.get_sensor().get_length() # heater line length for sensor
        self.temperature = (self.resistance-TCR[1])/TCR[0] # Calculate temperature of heater line from calibrated resistance thermometry [C]
        self.Trise3w = (2.0/(self.base.get_current()*TCR[0]))*np.array([self.base.get_v3wr()[self.min_freq_index], self.base.get_v3wr()[self.max_freq_index]]) # T-rise at lowest/highest fit frequency (Cahill eqn 12 w/ erratum)
        kSensorFit = self.base.get_sensor().get_KT_fit() # k fit for sensor [dk/dT,k-intercept] W/m^*K
        self.kSensor=np.polyval(kSensorFit, self.temperature) # calculate k of sensor from temperature for sensor 3
        
        # Calculate (total substrate + sample) thermal conductivities:
        self.kInPhase=(TCR[0]*self.resistance*self.base.get_current()**3)/(4*np.pi*L*self.line_fit[0]) # reminder: self.line_fit[0] is slope of fit
        self.kOutOfPhase=(TCR[0]*self.resistance*self.base.get_current()**3)/(8*self.fitted_v3wy_calc*L)
        if add_fit_to_plot:
            self.plot_fit() # plot the fit
        if save_figs:
            fig_name = self.base.get_full_fileName() + '_fit.pdf'
            self.base.get_figure().savefig(fig_name)
    def __str__(self):
        L = self.base.get_sensor().get_length()
        # section break between data files
        bl = 80 # "break length" (length of dividing lines separating files in output .txt file)
        trial = self.base.get_fileName()
        star_length = (bl-len(trial)-2)/2
        out_verb = '*'*star_length + ' ' + trial + ' ' + '*'*star_length + '*'*(len(trial)%2) + '\n'
        out_verb += '*'*bl + '\n'
        
        # General fit properties
        out_verb += '--------------- Supplied & Calculated Properties ---------------\n'
        out_verb += 'Out-of-Phase V3w avg rms value:   {0:.3e} Volts\n'.format(self.fitted_v3wy_calc)
        out_verb += 'In-Phase V3w Linear Fit:          Y = {0:.3e}*X + {1:.3e}\n'.format(self.line_fit[0],self.line_fit[1])
        out_verb += 'R-squared: {0:.4f}                 [Y] = Volts rms, [X] = ln(freq)\n'.format(self.R_squared)
        out_verb += 'Fitted to {0:.0f} frequency points.\n'.format(len(self.fitted_freq))
        
        # Sensor properties
        out_verb += '\nSensor:\n'
        out_verb += 'Calculated using:       {0:s}\n'.format(self.base.get_sensor().get_name())
        out_verb += 'Heater line length:     {0:.1f} mm\n'.format(L*1000)
        out_verb += 'Temp. Coeff. of Res:    {0:.5f} 1/K\n'.format(self.base.get_sensor().get_RT_fit()[0]/self.resistance)
        out_verb += 'Elec. resistance:       {0:.3g} Ohms\n'.format(self.resistance)
        out_verb += 'Thermal Conductivity:   {0:.4g} W/m*K\n'.format(self.kSensor)
        
        # Heating properties
        out_verb += '\nHeating (RMS values):\n'
        out_verb += 'Electrical current:     {0:.2f} mA\n'.format(1000*self.base.get_current())
        out_verb += 'Heating power:          {0:.3g} mW\n'.format(1000*self.base.get_current()**2*self.resistance)
        out_verb += 'Avg heater line T:      {0:.2f} C\n'.format(self.temperature)
        out_verb += 'Min fitted frequency:   {0:.3g} Hz, with 2-omega T-rise of {1:.2f} C\n'.format(np.min(self.fitted_freq), self.Trise3w[0])
        out_verb += 'Max fitted frequency:   {0:.3g} Hz, with 2-omega T-rise of {1:.2f} C\n'.format(np.max(self.fitted_freq), self.Trise3w[1])
        out_verb += 'Rough DC heating est:   {0:.2f} C\n'.format((self.base.get_current()**2*self.resistance/L)*np.log(.001/10E-6)/(np.pi*self.kInPhase))
        # Above is a very rough estimate of the temperature rise due to DC heating,
        # assuming the temperature drop is over ~ 1 mm, using radial cylindrical
        # SS heat conduction, assuming a heater line width of ~ 20 microns, and
        # assuming one infinite medium with an effective k the simple average of
        # the substrate and sample (i.e. kInPhase/2).
        out_verb += '------------------------------------------------------------------\n'
        
        # Print note
        out_verb += '\nNote recorded with data:\n'
        out_verb += self.base.get_comment() + '\n'
        
        # print (total combined) k values
        out_verb += '\n(out of Phase Total) k = \t{0:.4f} W/m*K\n'.format(self.kOutOfPhase)
        out_verb += '(In Phase Total) k = \t\t{0:.3f} W/m*K\n'.format(self.kInPhase)
        
        # Print sample k values
        out_verb += '\nk_sample = k_total - k_sensor:\n'
        out_verb += '(out of Phase Sample) k = \t{0:.3f} W/m*K\n'.format(self.kOutOfPhase-self.kSensor)
        out_verb += '(In Phase Sample) k = \t\t** {0:.3f} W/m*K **\n'.format(self.kInPhase-self.kSensor)
        
        # section break between data files
        out_verb += '\n' + '*'*bl +'\n'
        return out_verb
    def plot_fit(self):
        """ Select the appropriate figure associated with this trial
            and superpose (plot) this fit onto that figure """
        plt.figure(self.fig.number) # select the figure in which this data is plotted
        self.fig.add_subplot(1,2,1) # select the V1w sub-plot
        plt.axvline(x=self.fitted_freq[0],color='k', linewidth=1, label="Fit region " + str(self.fit_num)) # show left bound of fit inteval on v1w plot
        plt.axvline(x=self.fitted_freq[-1],color='k', linewidth=1) # show right bound of fit inteval on v1w plot
        plt.legend(loc="best")
        self.fig.add_subplot(1,2,2) # select the V3w sub-plot
        plt.plot(self.fitted_freq,self.fitted_v3wx_calc, 'k-', linewidth=3, label="Fit line " + str(self.fit_num)) # plot v3wx fit
        plt.plot(self.fitted_freq,self.fitted_v3wy_calc*np.ones(len(self.fitted_freq)), 'k-', linewidth=3) # plot v3wy fit
        plt.xlim(self.xlims_3w) # reset plot axes to original limits
        plt.ylim(self.ylims_3w)
        plt.legend(loc="best")
        
        # Annote figure with fit line equations / error
        xFitLabel = "V$_{3\omega ,x}$ = %.4g*ln($\omega$) + %.4g \nk$_{InPhase}$ = %.3f, R$^{2}$ = %.5g" % (self.line_fit[0], self.line_fit[1], self.kInPhase-self.kSensor, self.R_squared)
        yFitLabel = "V$_{3\omega ,y}$ = %.4g \nStd. Dev. = %.3g%%" % (self.fitted_v3wy_calc, 100*np.std(self.fitted_v3wy_exp)/self.fitted_v3wy_calc)
        V1wFitLabel = "Temperature$_{V1\omega}$ = %.1f $^{\circ}C$\nStd. Dev. = %.3g%%" % (self.temperature, 100*np.std(self.base.get_v1wr())/self.base.get_v1wr().mean())
        plt.annotate(xFitLabel, xy=(np.mean(self.fitted_freq), np.mean(self.fitted_v3wx_calc)), # annotate V3wx fit
                     xytext=(0.6,0.02), textcoords='figure fraction', fontsize=14, 
                     arrowprops=dict(facecolor='black', shrink=0.03, width=1), 
                     bbox=dict(facecolor="yellow", alpha=0.4, edgecolor="black", pad=5))
        plt.annotate(yFitLabel, xy=(np.mean(self.fitted_freq), self.fitted_v3wy_calc), # annotate V3wy fit
                     xytext=(0.7,0.9), textcoords='figure fraction', fontsize=14, 
                     arrowprops=dict(facecolor='black', shrink=0.05, width=1),
                     bbox=dict(facecolor="yellow", alpha=0.4, edgecolor="black", pad=10))
        plt.figtext(0.2,0.9,V1wFitLabel, fontsize=14, bbox=dict(facecolor="yellow", alpha=0.4, edgecolor="black", pad=10))
    
    
    
    def get_min_index(self): # get index for min fitted point
        return self.min_freq_index
    def get_max_index(self): # get index for max fitted point
        return self.max_freq_index
    def get_freq_fit(self): # return array of fitted frequencies [Hz]
        return self.fitted_freq
    def get_v3wy_fit(self): # return float of fitted v3wy value [Volts]
        return self.fitted_v3wy_calc
    def get_v3wx_fit(self): # return array of fitted v3wx values [Volts]
        return self.fitted_v3wx_calc
    def get_v1wr_fit(self): # return array of fitted v3wx values [Volts]
        return self.fitted_v1wr_calc
    def get_fit(self): # return array of [slope, intercept] for fit line
        return self.line_fit
    def get_R_squared(self): # retrn R-squared value for fit line
        return self.R_squared
    def get_temperature(self): # retrn R-squared value for fit line
        return self.temperature
    def get_kSensor(self): # retrn R-squared value for fit line
        return self.kSensor
    def get_kInPhase(self): # retrn R-squared value for fit line
        return self.kInPhase
    def get_kOutOfPhase(self): # retrn R-squared value for fit line
        return self.kOutOfPhase
    def get_verbose_output(self): # retrn R-squared value for fit line
        return self.out_verb


#==============================================================================
# Sensor Data Loading
#==============================================================================
class sensor(object):
    def __init__(self, temp_name, temp_RT_fit, temp_KT_fit, temp_length, temp_width, temp_thickness, temp_info):
        """ Generates a sensor object, storing associated data for a given sensor together.
        All arguments are given as lists of strings, except temp_name is just a single string """
        self.name = temp_name
        try:
            self.RT_fit = map(lambda x: float(x), temp_RT_fit)
        except ValueError:
            print 'WARNING: Separate coefficients in RTFIT of sensor "' + self.name + '" by spaces only'
        except:
            print 'WARNING: sensor "' + self.name +'" may be missing the RTFIT field; set to -1'
            self.RT_fit = -1
        try:
            self.KT_fit = map(lambda x: float(x), temp_KT_fit)
        except ValueError:
            print 'WARNING: Separate coefficients in KTFIT of sensor "' + self.name + '" by spaces only'
        except:
            print 'WARNING: sensor "' + self.name +'" may be missing the KTFIT field; set to -1'
            self.KT_fit = -1
        try:
            self.length = float(temp_length[0])
        except:
            print 'WARNING: sensor "' + self.name +'" may be missing the LENGTH field; set to -1'
            self.length = -1
        try:
            self.width = float(temp_width[0])
        except:
            print 'WARNING: sensor "' + self.name +'" may be missing the WIDTH field; set to -1'
            self.width = -1
        try:
            self.thickness = float(temp_thickness[0])
        except:
            print 'WARNING: sensor "' + self.name +'" may be missing the THICKNESS field; set to -1'
            self.thickness = -1
        try:
            self.info = " ".join(temp_info)
        except:
            self.info = ""
    def __str__(self):
        """ Specify the string representation of the Sensor object """
        strRep = "Sensor Name: " + self.name
        if type(self.RT_fit) == list:
            strRep += "\r RT_fit: " + str(self.RT_fit)
        if type(self.KT_fit) == list:
            strRep += "\r KT_fit: " + str(self.KT_fit)
        if self.length > 0:
            strRep += "\r Length: " + str(self.length*1e3) + " mm"
        if self.width > 0:
            strRep += "\r Width: " + str(self.width*1e6) + " microns"
        if self.thickness > 0:
            strRep += "\r Thickness: " + str(self.thickness*1e9) + " nm"
        if self.info != "":
            strRep += "\r Info: " + self.info + "\r"
        return strRep
    def get_name(self):
        return self.name
    def get_RT_fit(self):
        return self.RT_fit
    def get_KT_fit(self):
        return self.KT_fit
    def get_length(self):
        return self.length
    def get_width(self):
        return self.width
    def get_thickness(self):
        return self.thickness
    def get_info(self):
        return self.info
                
def get_sensors_list(sensors_file):
    """ Returns a list of trigger objectsnthat correspond to the rules set in the file filename """
    # Read in the sensor calibration file and eliminate blank lines and comments
    try:
        sensorsfile = open(sensors_file, "r")
    except:
        print 'There is no sensor calibrations data file named "' + sensors_file + '" in the directory'
        raise
    calibrations = [ line.rstrip() for line in sensorsfile.readlines() ]
    lines = [] # initialize lines list
    for line in calibrations:
        if len(line) == 0 or line[0] == '#': # remove blank lines & comments
            continue
        lines.append(line)
    """ Format:
    NAME name of sensor
    RTFIT coefficients of fit to electrical resistance vs temperature (typically linear), starting with coefficient of highest power [Ohms, K]
    KTFIT coefficients of fit to thermal conductivity vs temperature (typically linear) [W/m^2*K, K]
    LENGTH length of heater line [m]
    WIDTH width of heater line [m]
    THICKNESS thickness of heater line [m]
    INFO any other relevant info associated with that sensor, such as material, date created, use history, unique features, etc. """
    temp_name = None
    temp_RT_fit = None
    temp_KT_fit = None
    temp_length = None
    temp_width = None
    temp_thickness = None
    temp_info = None
    sensors = {} # initialize dictionary of sensors
    for line in lines:
        if line.split(" ")[0].strip(string.punctuation+" ") == "NAME":
            if temp_name != None:
                while temp_name in sensors: # don't overwrite an existing sensor:
                    print 'WARNING: sensor name "' + temp_name + '" appears more than once in file: "' + sensors_file + '"'
                    temp_name += " copy"
                sensors[temp_name] = sensor(temp_name, temp_RT_fit, temp_KT_fit, temp_length,
                                            temp_width, temp_thickness, temp_info)
            temp_name = None
            temp_RT_fit = None
            temp_KT_fit = None
            temp_length = None
            temp_width = None
            temp_thickness = None
            temp_info = None
            temp_name = " ".join(line.split(" ")[1:])
        if line.split(" ")[0].strip(string.punctuation+" ") == "RTFIT":
            temp_RT_fit = line.split(" ")[1:]
        if line.split(" ")[0].strip(string.punctuation+" ") == "KTFIT":
            temp_KT_fit = line.split(" ")[1:]
        if line.split(" ")[0].strip(string.punctuation+" ") == "LENGTH":
            temp_length = line.split(" ")[1:]
        if line.split(" ")[0].strip(string.punctuation+" ") == "WIDTH":
            temp_width = line.split(" ")[1:]
        if line.split(" ")[0].strip(string.punctuation+" ") == "THICKNESS":
            temp_thickness = line.split(" ")[1:]
        if line.split(" ")[0].strip(string.punctuation+" ") == "INFO":
            temp_info = line.split(" ")[1:]
    sensors[temp_name] = sensor(temp_name, temp_RT_fit, temp_KT_fit, temp_length,
                                            temp_width, temp_thickness, temp_info)
    return sensors
    
#==============================================================================
# Data Analysis
#==============================================================================
def analyze_all(data_directory, verbose_output=True, compact_output=True, 
                save_figs=False, perform_fit=True):
    """ Steps through every data file in the first level of 'data_directory' and runs a plot + fit protocol (does not look 
        at children directories).  Pickles (and returns) a list of the resulting fitted data_file objects, and saves two .txt 
        output files.  One verbose file, with full details of the analysis, another tab-delimited values suitable to 
        copy/paste into excel. """
    #import pickle
    
    out_strings = ['','',True] # format: [verbose, compact, write_headers]; verbose = description & full detail, compact = for copy/paste to excel
    
    # Define file-analysis protocol for a single data file
    def analyze_one(directory, trial, perform_fit=True):
        """ Analyze the 3-omega data file located at /directory/trial.  Parses data file, plots, applies fit, and saves
            a pickle, and both a verbose and compact output string of the results """
        file_path = directory + '/' + trial
        a = data_file(file_path)
        a.plot_data(selectFit=perform_fit)
        
        #-------------------------------------------------------------------------
        # Define data to be output to compact string. Format: [ (name, value), (name, value), ... ]
        compact_data = [('Heater Line Temperature [C]', a.get_fits()[0].get_temperature()), 
                        #('(combined) k In Phase [W/m*K]', kInPhase),
                        #('(combined) k Out of Phase [W/m*K]', kOutOfPhase),
                        ('(sample) k In Phase [W/m*K]', a.get_fits()[0].get_kInPhase()-a.get_fits()[0].get_kSensor()),
                        ('(sample) k Out of Phase [W/m*K]', a.get_fits()[0].get_kOutOfPhase()-a.get_fits()[0].get_kSensor()),
                        ('Min Freq [Hz]', np.min(a.get_fits()[0].get_freq_fit()) ),
                        ('Max Freq [Hz]', np.max(a.get_fits()[0].get_freq_fit()) ),
                        ('Num Freq Points', len(a.get_fits()[0].get_freq_fit()) ),
                        ('Data File Name', trial),
                        ('Sensor Used',a.get_sensor().get_name())]
        
        out_strings[0] += str(a.get_fits()[0])
        # Add data to compact string
        if out_strings[2]: # 3rd element set to True for first iteration, to write data column headers to file
            out_strings[2] = False
            out_strings[1] += '\t'.join([x[0] for x in compact_data]) + '\n'
        out_strings[1] += '\t'.join([str(x[1]) for x in compact_data]) + '\n'
        
    # Analyze each data file by having user select fitting frequency region, then store results
    if os.path.isdir(data_directory): # if user gave directory for mulitple data files
        pickle_name = data_directory.split('/')[-1] + '_fitted.p'
        pickle_dir = data_directory + '/' + pickle_name
        out_verbose_name = data_directory.split('/')[-1] + '_verbose.txt'
        out_verbose_dir = data_directory + '/' + out_verbose_name
        out_compact_name = data_directory.split('/')[-1] + '_compact.txt'
        out_compact_dir = data_directory + '/' + out_compact_name
        trials = os.listdir(data_directory)
        trials.sort(key = lambda x: x.split('_')[-1]) # group files by run number
        for trial in trials:
            if os.path.isfile(data_directory + '/' + trial):
                type_checker = trial.split('.')
                if (type_checker[0][:2] == 'S_') and ((type_checker[-1] in ['txt', 'csv']) or (len(type_checker) in [1, 2])):
                    plt.close('all') # leave the last plotted trial's fit open
                    analyze_one(data_directory, trial, perform_fit=perform_fit) # loop through and analyze each file
    elif os.path.isfile(data_directory): # user gave diretory for a single data file
        trial = data_directory.split('/')[-1]
        data_directory = '/'.join(data_directory.split('/')[:-1]) # chop off the last piece
        type_checker = trial.split('.')
        if (type_checker[0][:2] == 'S_') and ((type_checker[-1] in ['txt', 'csv']) or (len(type_checker) in [1, 2])):
            plt.close('all') # leave the last plotted trial's fit open
            if trial.split('.')[-1] in ['txt', 'csv']:
                out_verbose_name = '.'.join(trial.split('.')[:-1]) + '_verbose.txt'
                out_compact_name = '.'.join(trial.split('.')[:-1]) + '_compact.txt'
            else:
                out_verbose_name = trial + '_verbose.txt'
                out_compact_name = trial + '_compact.txt'
            out_verbose_dir = data_directory + '/' + out_verbose_name
            out_compact_dir = data_directory + '/' + out_compact_name
            analyze_one(data_directory, trial, perform_fit=perform_fit)
    
    # save verbose output string
    if verbose_output:
        with open(out_verbose_dir, "w") as text_file:
            text_file.write(out_strings[0])
    
    # save compact output string
    if compact_output:
        with open(out_compact_dir, "w") as text_file:
            text_file.write(out_strings[1])
```
